```python
# The values are presented in 2 formats:
# $79K-$147K (Glassdoor est.) & $145K-$225K(Employer est.)
# Hence, to convert this format to 79-147, 145-225, split salary into min_salary, max_salary, avg_salary

uncleaned_df["min_salary"]=0
uncleaned_df["max_salary"]=0
uncleaned_df["avg_salary"]=0

for i in range(len(uncleaned_df)):
    try:
        uncleaned_df.loc[i,"min_salary"]=int(uncleaned_df['Salary Estimate'][i].split(" ")[0].split("-")[0].replace("$","").replace("K",""))
        uncleaned_df.loc[i,"max_salary"]=int(uncleaned_df['Salary Estimate'][i].split(" ")[0].split("-")[1].replace("$","").replace("K",""))
    except:
        uncleaned_df.loc[i,"min_salary"]=int(uncleaned_df['Salary Estimate'][i].split("(E")[0].split("-")[0].replace("$","").replace("K",""))
        uncleaned_df.loc[i,"max_salary"]=int(uncleaned_df['Salary Estimate'][i].split("(E")[0].split("-")[1].replace("$","").replace("K",""))
    finally:
        uncleaned_df.loc[i,"Salary Estimate"]=str(uncleaned_df.loc[i,"min_salary"])+"-"+str(uncleaned_df.loc[i,"max_salary"])
        uncleaned_df.loc[i,"avg_salary"]=np.mean([uncleaned_df.loc[i,"min_salary"],uncleaned_df.loc[i,"max_salary"]])
```

```python
# simplifying job title
uncleaned_df['Job Title'].value_counts()[:100]
```

```python
def seniority(job_title):
    job_title=job_title.lower()
    snr=['sr','senior','lead','principal','mamager','vp','vice president','director']
    for i in snr:
        if i in job_title:
            return 'senior'
    if 'jr' in job_title:
        return 'junior'
    return 'na'
```

```python
uncleaned_df['seniority']=uncleaned_df['Job Title'].apply(seniority)
```

```python
# checking if seniority column has been applied correctly

uncleaned_df[uncleaned_df['Job Title']=='Vice President, Biometrics and Clinical Data Management']
```

```python
# simplifying job title

def simplified_title(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'machine learning'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    elif 'vice president' in title.lower():
        return 'vp'
    elif 'vp' in title.lower():
        return 'vp'
    else:
        return 'na'
uncleaned_df['job_simp']=uncleaned_df['Job Title'].apply(simplified_title)
```
